```python
from environment import FixedBudgetEnvironment
import matplotlib.pyplot as plt
import math
import numpy as np
from game import SumUnanimityGames, GlobalFeatureImportance, LocalFeatureImportance, UnsupervisedFeatureImportance
from plot import plot_fixedBudget
```

```python
from algorithms.CMCS import CMCS, Selective_CMCS
from algorithms.GapE import GapE
from algorithms.ApproShapley import ApproShapley
from algorithms.BUS import BUS
from algorithms.HybridApproBUS import SmartHybridApproBUS
from algorithms.SVARM import StratSVARM
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
n=16
K=np.arange(n-1)+1
budget=200
rounds=500
fixed_env = FixedBudgetEnvironment(n, budget)
results = []
```

```python
np.set_printoptions(formatter={'float': lambda x: "{0:0.10f}".format(x)})
```

```python
filepath_global = "datasets/Global feature importance/Bank marketing classification random forest.csv"
filepath_unsupervised = "datasets/Unsupervised Feature Importance/vf_Fifa.csv"
directory_local = "datasets/adult classification"
game = GlobalFeatureImportance(filepath=filepath_global, num_players=n, use_cached=True)
# game = LocalFeatureImportance(directory=directory_local, num_players=n)
# game = UnsupervisedFeatureImportance(filepath=filepath_unsupervised, num_players=n)
# game = SumUnanimityGames()
```

```python
results += [("Selective_CMCS", fixed_env.evaluate(game, Selective_CMCS(), K=K, rounds=rounds))]
```

```python
results += [("CMCS", fixed_env.evaluate(game, CMCS(), K=K, rounds=rounds))]
```

```python
results += [("StratSVARM(start_exact=True, theoretical_distribution=False)", fixed_env.evaluate(game, StratSVARM(start_exact=True, theoretical_distribution=False), K=K, rounds=rounds))]
```

```python
results += [("ApproShapley", fixed_env.evaluate(game, ApproShapley(optimize=True), K=K, rounds=rounds))]
```

```python
results += [("BUS", fixed_env.evaluate(game, BUS(), K=K, rounds=rounds))]
```

```python
results += [("Gap-E", fixed_env.evaluate(game, GapE(), K=K, rounds=rounds))]
```

```python
plot_fixedBudget(results, K, save=True, filepath=f"results/fixed_budget_{game.__class__.__name__}({game.name})_n={n}_T={budget}_r={rounds}.pdf")
```
